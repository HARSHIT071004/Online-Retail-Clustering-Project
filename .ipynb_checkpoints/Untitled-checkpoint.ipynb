`Outliers Checking`

```python
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col} → Outliers: {outliers.shape[0]}")

```

```python
# Outliers - fill it with median >> it is better choice than mean
```

```python

# Numeric columns को filter कर लो
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median = df[col].median()
    
    df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound),
                       median,
                       df[col])

```

`Key Insights and Observations`
1. Data Cleaning
Removed rows with negative or zero Quantity and UnitPrice values.

Converted InvoiceDate to datetime and extracted Year, Month, Hour, and Weekday.

Created TotalPrice column (Quantity * UnitPrice) for revenue-based analysis.

2. Sales Performance
Sales peak in November and December, indicating strong year-end seasonality.

Hourly analysis shows highest activity between 10 AM and 1 PM.

Thursday and Wednesday observed as top-performing weekdays.

3. Product-Level Insights
A small set of products contribute disproportionately to total revenue.

Top 10 products dominate overall sales volume and value.

4. Customer Behavior
Many customers show repeat purchases, indicating loyalty patterns.

Most purchases involve 1–5 items per invoice; few high-quantity orders exist.

5. Geographic Distribution
United Kingdom is the dominant market in terms of sales volume and revenue.

Other notable contributors include Germany, Netherlands, and France.

6. Invoice-Level Patterns
Total sales per invoice follow a right-skewed distribution — majority are low-value transactions.

Sales frequency throughout the day suggests consistent customer activity.

Conclusion:
The dataset shows strong seasonality, time-based trends, and country-level variation. The data is now clean and ready for clustering, RFM segmentation, or predictive modeling.
